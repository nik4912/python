```python
import pandas as pd

df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Mumbai', 'Delhi'],
    'Education': ['Graduate', 'PostGrad', 'School', 'Graduate', 'PostGrad'],
    'Purchased': [1, 0, 1, 0, 1]
})

print(df)
```

One-Hot Encoding(Pandas Method)
1) no order
2) avoid dummy data

```python
# no order 
df_ohe = pd.get_dummies(df, columns=['Gender', 'City'])

print(df_ohe)
```

```python
# avoid dummy
df_ohe = pd.get_dummies(df, columns=['Gender', 'City'], drop_first=True)
print(df_ohe)
```

One-Hot Using Sklearn 

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')

encoded = ohe.fit_transform(df[['Gender', 'City']])

encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())

print(encoded_df)
```
