```python
import pandas as pd

df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Mumbai', 'Delhi'],
    'Education': ['Graduate', 'PostGrad', 'School', 'Graduate', 'PostGrad'],
    'Purchased': [1, 0, 1, 0, 1]
})

print(df)
```

```python
df_target = df.copy()

target_mean = df_target.groupby('City')['Purchased'].mean()

df_target['City_encoded'] = df_target['City'].map(target_mean)

print(df_target)
```
