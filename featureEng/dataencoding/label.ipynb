```python
import pandas as pd

df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Mumbai', 'Delhi'],
    'Education': ['Graduate', 'PostGrad', 'School', 'Graduate', 'PostGrad'],
    'Purchased': [1, 0, 1, 0, 1]
})

print(df)
```

Label Encoding - Using Sklearn

Male/Female -> Convert in 0/1 

```python
from sklearn.preprocessing import LabelEncoder

# Create copy
df_label = df.copy()

le = LabelEncoder()

# Apply on Gender
df_label['Gender_encoded'] = le.fit_transform(df_label['Gender'])

print("Classes:", le.classes_)   # Shows mapping
print(df_label)
```

Label Encoding (Without Sklearn)

```python
df_manual = df.copy()

gender_mapping = {
    'Male': 1,
    'Female': 0
}

df_manual['Gender_encoded'] = df_manual['Gender'].map(gender_mapping)

print(df_manual)
```
